# VAE for Bird Phylogenetic Tree Construction

This notebook demonstrates how to use a Variational Autoencoder (VAE) to learn meaningful representations of bird spectrograms and construct phylogenetic trees from these representations.

## Overview
1. **Data Loading**: Load preprocessed spectrograms
2. **VAE Training**: Train a VAE to learn compressed representations
3. **Phylogenetic Analysis**: Use latent representations to construct phylogenetic trees
4. **Visualization**: Visualize the learned representations and phylogenetic relationships
